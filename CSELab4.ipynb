```python
#Password_code
password = input("Enter Your Password: ")
up = False
low = False
num = False
special = False

for c in password:
    if "A" <= c <= "Z":
        up = True
    elif "a" <= c <= "z":
        low = True
    elif "0" <= c <= "9":
        num = True
    elif c in "!@#$%^&*":
        special = True

if up and low and num and special==True:
    print("Your Password is valid")
else:
    print("Your Password is invalid")
```

```python
#Task-6(Password)
u = "P@ssWord123"
sym = "!@#$%^&"
num = "0123456789"
symC = 0
numC = 0
uC = 0
lC = 0
uS = ""
lS = ""
symS = ""
numS = ""

for i in range(0, len(u)):
    if ord(u[i]) >= ord("A") and ord(u[i]) <= ord("Z"):
        uC += 1
        uS += u[i]
    elif ord(u[i]) >= ord("a") and ord(u[i]) <= ord("z"):
        lC += 1
        lS += u[i]
    elif u[i] in sym:
        symC += 1
        symS += u[i]
    elif u[i] in num:
        numC += 1
        numS += u[i]

if symC > 0 and numC > 0 and uC > 0 and lC > 0 and len(u) > 6:
    print("Strong Password")
else:
    if symC == 0:
        print("Symbol Missing")
    if numC == 0:
        print("Number Missing")
    if uC == 0:
        print("Upper Missing")
    if lC == 0:
        print("Lower Missing")
    if len(u) <= 6:
        print("Not Length")

print(uS, lS, symS, numS)
```

```python
#TAsk-19
u = "Python programming is very easy"
prev = False  #False maneh lower
n = ""

for curr in u:
    if prev == False:
        if ord(curr) >= ord("A") and ord(curr) <= ord("Z"):
            n += curr
            prev = True
        elif ord(curr) >= ord("a") and ord(curr) <= ord("z"):
            convert = ord(curr) - 32
            n += chr(convert)
            prev = True
        else:
            n += curr
    elif prev == True:
        if ord(curr) >= ord("a") and ord(curr) <= ord("z"):
            n += curr
            prev = False
        elif ord(curr) >= ord("A") and ord(curr) <= ord("Z"):
            convert = ord(curr) + 32
            n += chr(convert)
            prev = False
        else:
            n += curr

print(n)
```
