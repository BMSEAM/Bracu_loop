```python
class Match:
  def __init__(self, *args):
    if len(args) == 2:
      self.run = args[0]
      self.wicket = args[1]
      self.team = 'BD'

    elif len(args) == 3:
      self.team = args[0]
      self.run = args[1]
      self.wicket = args[2]

  def show(self):
    s = f'Team {self.team}. Score: {self.run}/{self.wicket}'
    return s


#Driver Code
b1 = Match(120,5) #-->0th run 1st wicket
print(b1.show())
print('---------------')
b1 = Match('IND',189,3) #0th name 1st run 2nd wicket
print(b1.show())
```

```python
class Department:
  def __init__(self, s):
    self.semester = s
    self.name = "Default"
    self.id = -1

  def student_info(self):
    print("Name:", self.name)
    print("ID:", self.id)

  def courses(self, *args):
    print("No courses Approved yet!")

  def methodA(self):
    print(self.id)

class CSE (Department):
  def __init__(self, name, id, semester):
    super().__init__(semester) #semster = Spring2016, name = 'Default', id = -1
    self.name = name #name = 'Rahim'
    self.id = id #id = 16101328
    #we are overrriding the intstance variables of parent class - name, id
    #overwrite == overrride
    self.course_list = []

#we are overrriding the intstance method of parent class - courses
  def courses(self, *args):
    if len(args) == 0:
      super().courses()
    else:
      self.course_list += list(args)

  # def methodA(self):
  #   super().methodA()

#we are overrriding the intstance method of parent class - student_info
  def student_info(self):
    super().student_info()
    print(f'Courses Approved to this CSE student in {self.semester} semester :')
    for i in self.course_list:
      print(i)





s1 = CSE("Rahim", 16101328,"Spring2016")
s1.courses()
s1.methodA()
print("==================")
s1.courses("CSE110", "MAT110")
s1.student_info()
print("==================")
s1.courses("ENG101")
s1.student_info()
print("==================")
s2 = EEE("Tanzim", 18101326, "Spring2018")
s2.courses("Mat110", "PHY111", "ENG101")
s2.student_info()
print("==================")
s3 = CSE("Rudana", 18101326, "Fall2017")
s3.courses("CSE111")
s3.student_info()
print("==================")
s4 = EEE("Zainab", 19201623, "Summer2019")
s4.courses("EEE201", "PHY112", "MAT120")
s4.student_info()
```
